```python
# creating a dictionary that contains three dictionary
family = {
    "ram":{
        "role":"dad",
        "age":"50",
        "responsibility":"earn money and look after his family"
    },
    "sita":{
        "role":"mother",
        "age":"48",
        "responsibility":"household works and look after her children"
    },
    "timin":{
        "role":"children",
        "age":18,
        "responsibility" : "study and work hard"
    }
}
print(family)
print(family["ram"])
print(family["ram"]["age"])
```

```python
#another way of making nested dictionary
ram = {
        "role":"dad",
        "age":"50",
        "responsibility":"earn money and look after his family"
    }
sita = {
        "role":"mother",
        "age":"48",
        "responsibility":"household works and look after her children"
    }

timin = {
        "role":"children",
        "age":18,
        "responsibility" : "study and work hard"
    }

family = {
    "ram": ram,
    "sita":sita,
    "timin":timin
}
print(family)
print(family["ram"])
print(family["ram"]["age"])
```

```python
for key, value in family.items():
    print(f"key = {key}, value = {value}")
    for y,y_value in value.items():
        print(f"key = {y}, value = {y_value}")
```
