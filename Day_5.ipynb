````python
print("Looping from 0 to 4:")
for i in range(5): # 0, 1, 2, 3, 4
    print(i)

print("-" * 20)
````

````python
print("Looping from 2 to 8:")
for i in range(2,8): 
    print(i)

print("-" * 20)
````

````python
message = "Hello"
print("Looping through a string:")
for char in message:
    print(char)

print("-" * 20)
````

````python
fruits = ["apple", "banana", "cherry"]

print("Looping through fruits with index:")
for i,fruit in enumerate(fruits):
    print(f" {i+1}: {fruit}")
````

`while` loops execute a block of code *as long as* a specified condition remains `True`.

**Syntax:**

```python
while a>10:
    # Code block to execute as long as condition is True
    print("Condition is still True")
    # IMPORTANT: Usually, something inside the loop needs to
    # change the condition, otherwise it might loop forever!

````python
count = 0
while count<10:
    print(f"count = {count}")
    count += 1
````
