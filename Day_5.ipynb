## 3. Loop Control Statements (`break`, `continue`, `pass`)

These statements allow you to modify the standard behavior of loops.

*   **`break`**: — Stop the loop right now!
*   **`continue`**: Skip this one and move on.
*   **`pass`**: Does nothing. It acts as a placeholder where syntax requires a statement, but you don't want any code to execute. This is often used in empty function definitions, classes, or conditional blocks during development.

````python
# Example 1: break
print("Finding the first multiple of 7 between 1 and 20:")
list1 = []
for number in range(1, 21):
   
    if number % 7 == 0:
        print(f"Found it! {number} is a multiple of 7.")
        break
    else:
        list1.append(number)

print("Loop finished or broken.")
print("Lists:", list1)
````

````python
# Example 2: continue
print("Finding the multiple of 7 between 1 and 22:")
lists = []
for number in range(1, 22):
   
    if number %7  == 0:
        print(f"Found it! {number} is a multiple of 7.")
        continue
    else:
        lists.append(number)

print("Loop finished or broken.")
print("Lists:", lists)
````

````python
# Example : break(finds the first multiple of 14)
print("Finding the first multiple of 14 between 1 and 20:")
list1 = []
for number in range(1, 21):
   
    if number % 14 == 0:
        print(f"Found it! {number} is a multiple of 14.")
        break
    else:
        list1.append(number)

print("Loop finished or broken.")
print("Lists:", list1)
````

````python
# Example 2: Using 'continue' to print only even numbers
print("Printing even numbers between 1 and 10:")
for number in range(1, 11):
    if number % 2 != 0:
        continue

    print(f"Even number: {number}")

print('Loop finished\'s.')
````

### Activity 1: Number Guessing Game

**Goal:** The computer picks a secret number, and the user has to guess it. The program gives hints like "too high" or "too low".

**Concepts We'll Use:**
*   `import random` (to get a random number)
*   `input()` (to get the user's guess)
*   `int()` (to convert the user's input string into a number)
*   `while` loop (to keep guessing until correct)
*   `if`/`elif`/`else` (to check the guess and give hints)

````python
## import random
secret_num = random.randint(1,20)
print(f"{secret_num}")
user_input = int(input("Guess a number:"))
while True:
    if user_input > secret_num:
        print("You guess a number which is high")
    elif user_input < secret_num:
        print("You guess a number which is low")
    else:
        print("Congratulations! You guess the correct number.")
        break
    user_input = int(input("Guess a number:"))
    
    
````
