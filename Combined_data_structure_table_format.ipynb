```python
# practice
# total marks of each student
marks = [
    [85, 90, 88],
    [78, 82, 80],
    [92, 95, 94]
]
count = 0

for item in marks:
    total = 0
    for mark in item:
        total = total + mark
    count +=1 
    print(f"Total marks of student {count} = {total}")
    
```

```python
# practice
# find highest mark
marks = [
    [85, 90, 88],
    [78, 82, 80],
    [92, 95, 94]
]
max_num = marks[0][0]
for item in marks:
    for mark in item:
        if max_num < mark:
            max_num = mark
print(f"The highest mark in the list marks = {max_num}")
```

```python
# find average of marks
student = {
    "name": "Sarita",
    "marks": [85, 90, 88]
}
print(f"Name: {student["name"]}")
print(f"Marks:\n{student["marks"][0]}\n{student["marks"][1]}\n{student["marks"][2]}")
total = 0
for mark in student["marks"]:
    total = total + mark
avg = total/len(student["marks"])
print(len(student["marks"]))
print(avg)    
```

```python
# list of dictionaries (find highest mark and topper )
students = [
    {"name": "Ram", "marks": 80},
    {"name": "Shyam", "marks": 75},
    {"name": "Hari", "marks": 90}
]
high_mark = students[0]["marks"]
for student in students:
    print(f"{student["name"]} -> {student["marks"]}")
    if high_mark<student["marks"]:
        high_mark = student["marks"]
for student in students:
    if high_mark== student["marks"]:
        print(f"Topper = {student["name"]}")
        print(f"Marks = {student["marks"]}")
    
```

```python
# dictionary of dictionaries
students = {
    "Ram": {
        "age": 20,
        "marks": 80
    },
    "Shyam": {
        "age": 21,
        "marks": 75
    },
    "Hari": {
        "age": 19,
        "marks": 90
    }
}

for student_name,student_info in students.items():
    print(f"{student_name}")
    print(f"Age: {student_info["age"]}\nMarks: {student_info["marks"]}")
    print("\n")
```

```python
# Student Result Management System
students = []
for i in range(3):
    name = input("Enter name:")
    age = int(input("Enter age:"))
    marks = int(input("Enter marks:"))
    students.append({"name":name,"age":age,"marks":marks})
    print()
for student in students:
    print(f"Name : {student["name"]}")
    print(f"Age : {student["age"]}")
    print(f"Marks : {student["marks"]}")
    print()
high_mark = students[0]["marks"]
for student in students:
    if high_mark < student["marks"]:
        high_mark = student["marks"]
total = 0
for student in students:
    if high_mark == student["marks"]:
        print(f"Student with highest marks = {student["marks"]}")
    total = total + student["marks"]
avg = total/3
print(f"Average marks of 3 students = {avg}")
```

```python
# Mini Employee Database
employees = {}
high_salary = 0
for i in range(3):
    employee_id = int(input("Enter employee ID:"))
    name = input("Enter name:")
    salary = int(input("Enter salary:"))
    department = input("Enter department:")
    employees.update({employee_id : {"name":name, "salary": salary, "department":department}})
    if high_salary< salary:
        high_salary = salary

print("1. Display all employees")
print("2. Search employee by ID")
print("3. Find employee with highest salary")
print("4. Exit")
while(True):
    choice = int(input("Enter your choice:"))
    if choice == 1:
        for employee_id, employee_details in employees.items():
            print(f"Employee_id = {employee_id} \n Name: {employee_details["name"]} \nSalary: {employee_details["salary"]}\nDepartment: {employee_details["department"]}")
    
    elif choice == 2:
        id = int(input("Enter employee ID:"))
        if id in employees.keys():
            print("Employee exists.")
    elif choice == 3:
        for employee_details in employees.values():
            if high_salary == employee_details["salary"]:
                print(f"Employee with high salary:\n Name : {employee_details["name"]} \n Salary : {employee_details["salary"]} \n Department : {employee_details["department"]}")
    elif choice == 4:
        print("Exiting..")
        break
    else:
        print("Invalid input.")
```
