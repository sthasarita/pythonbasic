3. Build a number classifier: positive/negative/zero,
odd/even

```python
print("Number Classifier Classifying positive, negative or zero")
number = int(input("Enter any number:"))
if number >0:
    print(f"{number} is positive.")
elif number < 0:
    print(f"{number} is negative.")
else:
    print(f"{number} is zero.")

print("Number Classifier Classifying odd or even")
number = int(input("Enter any number:"))
if number % 2 == 0:
    print(f"{number} is even.")
else:
    print(f"{number} is odd.")

```

4. Write a ticket pricing system: child (<10), student
(10-18), adult (18-60), senior (60+)

```python
print("---Ticket Pricing System---")
age = int(input("Enter your age:"))
if age<10:
    print(f"Category = Child \nTicket price is 50")
elif age >= 10 and age <= 18:
    print(f"Category = Student \nTicket price is 100")
elif age >= 18 and age <= 60:
    print(f"Category = Adult \nTicket price is 200")
else:
    print(f"Category = Senior \nTicket price is 100")
```

5. Create a simple ATM simulator: check balance, warn
if low

```python
print("-----NIC ASIA ATM-----")
print("-------Welcome--------")
balance = 50000
print("1. Check Balance")
print("2. Withdraw")
print("3. Deposit")
print("4. Exit")
choice = int(input("Enter your choice:"))
while True:
    if choice == 1:
        print(f"Balance = {balance}")
    elif choice == 2:
        withdraw_amt = int(input("Enter amount to withdraw:"))
        if withdraw_amt > balance:
            print("Sorry! You have insufficient balance.")
        else:
            balance = balance - withdraw_amt 
            print("Withdraw Success.Take your money.")
            print(f"Your new balance is {balance}")
    elif choice == 3:
        deposit_amt = int(input("Enter amount to deposit:"))
        balance = balance + deposit_amt
        print(f"Your new balance is {balance}")
    elif choice == 4:
        print("Exiting....")
        break
    else:
        print("Invalid input.")
    choice = int(input("Enter your choice:"))


        
    
```

8. Write a loop that finds all prime numbers between 1
and 100

```python
print("Prime numbers between 1 and 100-")
for i in range(1,101):
    num = i
    j = 2
    while j < num:
        if num % j == 0:
            break
        j += 1 
    if num == j :
        print(num, end = " ")
```

9. Build a simple number-guessing game using while
loop

```python
print("Number-guessing game-")
number = 4
guess_num = 2
while guess_num != number:
    guess_num = int(input("Guess a number:"))
    if guess_num == number:
        print("Congratulations! Your guess is correct.")
    else:
        print("Oops! Your guess is wrong.\nTry again.")
```
