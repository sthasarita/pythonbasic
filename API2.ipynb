```python
import requests

api_url = "https://jsonplaceholder.typicode.com/todo/1"

# Make the GET request
response = requests.get(api_url)
print(response)
# The response object contains all the information
print(f"Status Code: {response.status_code}") # 200 means OK

# The actual data is in the response body
# .text gives the raw text, .json() parses it directly into a Python dict
if response.status_code == 200:
    data = response.json()
    print("Data from API:")
    print(data)
    print(f"Title: {data['title']}")
else:
    print(f"Failed to fetch data. Status code: {response.status_code}")

    
```

```python

2. Parsing JSON from an API
Most modern APIs return data in JSON format. The response.json() method from the requests library makes it incredibly easy
to work with this data. It automatically converts the JSON string into a Python dictionary or list.
```

```python
import requests

# This API returns a list of posts
api_url = "https://jsonplaceholder.typicode.com/posts"

response = requests.get(api_url)

if response.status_code == 200:
    posts = response.json()
    print(f"Received {len(posts)} posts. Here are the first two:")
    
    # posts is a list of dictionaries
    for post in posts[:29]:
        print("---")
        print(f"User ID: {post['userId']}")
        print(f"Title: {post['title']}")
        print(f"Body: {post['body'][:30]}...") 
else:
    print("Failed to get posts.")

    
```

```python
API Error Handling
Status Codes: Always check response.status_code.
2xx (e.g., 200): Success
4xx (e.g., 404 Not Found, 401 Unauthorized, 429 Too Many Requests): Client-side error
5xx (e.g., 500 Internal Server Error): Server-side error

Exceptions: Network issues can raise exceptions. It's good practice to wrap API calls in a try-except block.
```

```python
import requests

try:
    # A URL that will likely give a 404 error
    response = requests.get("https://api.github.com/users/nonexistentuser123456789")
    
    # This will raise an HTTPError for 4xx or 5xx status codes
    response.raise_for_status() 

    data = response.json()
    print(data)

except requests.exceptions.HTTPError as errh:
    print(f"Http Error: {errh}")
except requests.exceptions.ConnectionError as errc:
    print(f"Error Connecting: {errc}")
except requests.exceptions.Timeout as errt:
    print(f"Timeout Error: {errt}")
except requests.exceptions.RequestException as err:
    print(f"Oops: Something Else: {err}")
    
```

```python
Mini-Project: Real-Time Currency Converter
Let's build a script that fetches live currency exchange rates and converts an amount from one currency to another. 
We'll use the free ExchangeRate-API.
```

```python
import requests

def get_exchange_rates(base_currency):
    """Fetches exchange rates from the API for a given base currency."""
    api_url = f"https://open.er-api.com/v6/latest/{base_currency}"
    try:
        response = requests.get(api_url)
        response.raise_for_status()
        data = response.json()
        if data.get('result') == 'success':
            print(data['rates'])
            return data['rates']
        else:
            print("API did not return a successful result.")
            return None
    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")
        return None

def convert_currency(amount, from_currency, to_currency, rates):
    """Converts an amount from one currency to another using the provided rates."""
    if not rates or to_currency not in rates:
        print(f"Conversion rate for {to_currency} not found.")
        return None
    
    rate = rates[to_currency]
    converted_amount = amount * rate
    return converted_amount

# --- Main script ---
base = "EUR"
target = "NPR"
amount_to_convert = 1

print(f"Fetching latest rates with base currency {base}...")
exchange_rates = get_exchange_rates(base)

if exchange_rates:
    converted_value = convert_currency(amount_to_convert, base, target, exchange_rates)
    if converted_value is not None:
        print(f"{amount_to_convert} {base} is equal to {converted_value:.2f} {target}")

```
