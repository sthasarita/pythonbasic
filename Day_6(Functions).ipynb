```python
def add(a,b):
    result = a+b
    result1 = a-b
    return result,result1

x = 10
y=4
add_result , sub_result = add(x,y) #unpacking
print(add_result,sub_result)
```

```python
def add(a,b):
    result = a+b
    result1 = a-b
    ignore = a
    return result,result1,ignore

x = 10
y=4
add_result , sub_result,_ = add(x,y) #unpacking
print(add_result,sub_result)
print(_)
```

```python
# *args and **kwargs
# Use *args to pass multiple positional arguments.


def show_args(*args):
   
    lists =""
    for item in args:
        print(item)
        lists += item
    return lists

value =show_args("111", "222")
print(value)
```

```python

def show_args(*args):
   
    lists =0
    for item in args:
        print(item)
        lists += item
    return lists

value =show_args(111, 222)
print(value)
```

```python
sum1 = 0
def addition(*args):
    for item in args:
        global sum1
        sum1 = sum1 + item
    return sum1

print(addition(1,2,4,22,44,55,55))
print(sum1)
```

```python
sum1 = 0
def addition(*args):
    for item in args:
        sum1 = 0
        sum1 = sum1 + item
    return sum1

print(addition(1,2,4,22,44,55,55))
print(sum1)
```

```python
#use **kwargs to pass multiple keyword arguments
name = input("Enter your name:")
age = input("Enter age:")
def show_kwargs(**kwargs):
    print(kwargs)
show_kwargs(name=name , age = age)
```

```python
name = input("Enter your name:")
age = input("Enter your age:")
def show_kwargs(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}= {value}")
show_kwargs(name = name , age = age , surname = "Shrestha", grade = "10")
```
