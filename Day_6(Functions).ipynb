```python
def add(a,b):
    result = a+b
    result1 = a-b
    return result,result1

x = 10
y=4
add_result , sub_result = add(x,y) #unpacking
print(add_result,sub_result)
```

```python
def add(a,b):
    result = a+b
    result1 = a-b
    ignore = a
    return result,result1,ignore

x = 10
y=4
add_result, sub_result , _= add(x,y) #unpacking
print(add_result,sub_result,_)
```

## 4. Lambda Functions (Anonymous Functions)

Python allows you to create small, unnamed functions using the `lambda` keyword. These are useful for short operations where defining a full function with `def` feels overly verbose.

**Syntax:**
`lambda arguments: expression`

*   `lambda`: Keyword.
*   `arguments`: Comma-separated arguments
*   `expression`: A single expression whose result is implicitly returned. Lambda functions cannot contain multiple statements or complex logic.

They are often used in places where a function object is required, like arguments to `sorted()`, `map()`, `filter()`.

```python
#using lambda function
double = lambda x,y: x * y
print(double(5,6))
print(type(double))
```

```python
#using normal function 
def double(x,y):
    return x*y
print(double(5,6))
print(type(double))
```

```python
double = lambda x,y: x * 2,y*2
print(double(5,6))
print(type(double))
```

```python
# adding two numbers using lambda functions
double = lambda x,y: x + y
print(double(5,6))
print(type(double))
```

```python
# sorting the list in ascending order of marks
students = [
    {"name": "Ram", "marks": 80},
    {"name": "Hari", "marks": 95},
    {"name": "Shyam", "marks": 75},
    {"name": "Gita", "marks": 88}
]
sorted_students = sorted(students, key = lambda x :x["marks"])
for student in sorted_students:
    print(f"{student["name"]} -> {student["marks"]}")
```
