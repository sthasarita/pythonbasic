## 3. Check Even or Odd
Write a program to check whether a given number is even or odd.

```python
n = int(input("Enter a number:"))
if n % 2 == 0:
    print(f"{n} is even.")
else:
    print(f"{n} is odd.")
```

## 4. Find the Largest Number
Write a program to find the largest among three numbers entered by the user.

```python
num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))
num3 = int(input("Enter third number:"))
if num1 > num2 and num1 > num3:
    print(f"{num1} is largest among {num1}, {num2} and {num3}")
elif num2 > num3 and num2 > num1:
    print(f"{num2} is largest among {num1}, {num2} and {num3}")
else:
    print(f"{num3} is largest among {num1}, {num2} and {num3}")
```

## 5. Count Vowels in a String 
Write a program to count the number of vowels in a given string.

```python
# Your code here
string = input("Enter a string:")
count = 0
for char in string:
    if char in ['a','e','i','o','u']:
        count += 1
print(f"No. of vowels in {string} = {count}")

```

## 6. Reverse a String 
Write a program to reverse a given string.

```python
# Your code here
string = input("Enter a string:")
rev_str = ""
for ch in string:
    rev_str = ch + rev_str
print(rev_str)
    
```

## 7. Factorial of a Number 
Write a program to calculate the factorial of a number using a loop.

```python
# Your code here
n = int(input("Enter a number:"))
f = 1
for i in range(1,n+1):
    f = f*i
print(f"Factorial of {n} is {f}")
```

## 8. Sum of List Elements
Find the sum of all elements in a list.

```python
# Your code here
li = [1,2,3]
total = 0
for n in li:
    total += n
print(f"Sum = {total}")
    
```

## 9. Find Maximum Element in a List

```python
# Your code here
number = [3,5,6,7]
maximum = number[0]
for n in number:
    if maximum < n:
        maximum= n 
print(f"Maximum = {maximum}")
    
```
