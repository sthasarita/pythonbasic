## Common Regex Functions & Outputs

Here is a breakdown of the most common `re` functions, what they do, and what type of data they return.

| Function | Simple Description | Return Output Type |
| :--- | :--- | :--- |
| **`re.search()`** | Scans the **whole text** for the **first** match. | `Match Object` (or `None`) |
| **`re.match()`** | Checks for a match **only at the very start** of the text. | `Match Object` (or `None`) |
| **`re.findall()`** | Finds **all** matches and puts the text in a list. | `List` of Strings |
| **`re.finditer()`** | Finds **all** matches but keeps them as objects (useful for locations). | `Iterator` (loopable) |
| **`re.sub()`** | Finds a pattern and **replaces** it with new text. | `String` (The new text) |
| **`re.split()`** | **Cuts** the string wherever the pattern matches. | `List` of Strings |

> **Note:** A `Match Object` contains details like the specific text found and its position (index). A `List` just contains the text.

```python
import re

text = "The rain in Spain falls mainly on the plain."
pattern = r"ain"


print(f"Original Text: '{text}'\n")

# 1. re.search()
res_search = re.search(pattern, text)
print(f"1. Search:   {res_search}")
```

```python
res_search.group(0)
```

```python
res_search.span()
```

```python
res_match = re.match(pattern, text[5:])
print(f"2. Match: {res_match}")
```

```python
res_findall = re.findall(pattern,text)
print(f"3. Findall :{res_findall}")
```

```python
print(len(res_findall))
print(type(res_findall))
```

```python
res_iter = re.finditer(pattern,text)

print(f"4. Finditer: {[m.group() for m in res_iter]}")
print(res_iter)
```

```python
res_iter = re.finditer(pattern,text)
print(list(res_iter))
print(f"4. Finditer: {[m.group() for m in res_iter]}")
print(res_iter)
```

```python
res_sub = re.sub(pattern, "YYY", text)
print(f"5. Sub: {res_sub}")
```

```python
res_split = re.split(pattern,  text)
print(f"5. split: {res_split}")
```

## Regex Metacharacter Cheatsheet

These special characters act as the building blocks for creating search patterns.

| Character | Description | Example Pattern | Matches Example |
| :---: | :--- | :--- | :--- |
| **`.`** | Any character (except newline) | `h.t` | hat, hot, hit |
| **`^`** | Starts with | `^Hello` | "Hello world" |
| **`$`** | Ends with | `world$` | "Hello world" |
| **`*`** | Zero or more times | `a*b` | b, ab, aab |
| **`+`** | One or more times | `a+b` | ab, aab |
| **`?`** | Zero or one time (Optional) | `colou?r` | color, colour |
| **`{}`** | Specific quantity range | `a{2,4}` | aa, aaa, aaaa |
| **`[]`** | A set of allowed characters | `[a-z]` | Any lowercase letter |
| **`\`** | Escape (Treat next char as normal) | `\.` | A literal dot "." |
| **`\|`** | OR operator | `cat\|dog` | cat, dog |
| **`()`** | Capture group | `(a\|b)c` | ac, bc |

```python
| Pattern   | Text       | Match? | Why                                              |
| --------- | ---------- | ------ | ------------------------------------------------ |
| `\bcat\b` | `cat`      | ✔      | Whole word                                       |
| `\bcat\b` | `category` | ✘      | `cat` is inside a larger word                    |
| `cat\B`   | `category` | ✔      | Another word character follows `cat`             |
| `cat\B`   | `cat `     | ✘      | Word ends after `t`                              |
| `\Bcat`   | `scat`     | ✔      | Before `c` is `s` (still inside a word)          |
| `\Bcat`   | ` cat`     | ✘      | Before `c` is a space, so that's a word boundary |
```

```python
Inside square brackets [], metacharacters like * or + lose their special meaning and are treated as literals.
[abc]: Matches 'a', 'b', or 'c'.
[a-z]: Matches any lowercase letter.
[A-Z]: Matches any uppercase letter.
[0-9]: Matches any digit.
[^abc]: Matches any character EXCEPT 'a', 'b', or 'c' (Negation).
```

```python
import re

text = "John Doe: 555-1234"
pattern = r"(\w+) (\w+): (\d{3}-\d{4})"
match = re.search(pattern, text)

if match:
    print(match.group(0)) # The entire match: "John Doe: 555-1234"
    print(match.group(1)) # First group: "John"
    print(match.group(2)) # Second group: "Doe"
    print(match.group(3)) # Third group: "555-1234"
```

```python
Named Groups: (?P<name>...):
Allows you to access groups by name instead of index
```
