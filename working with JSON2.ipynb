JavaScript Object Notation (JSON), one of the most common data formats used for `data 
exchange on the web`. We will learn its syntax, how to work with it in Python, and how 
to handle complex, nested data structures.

##### Learning Objectives
* Understand the structure and syntax of JSON.
* Serialize Python objects into JSON strings.
* Deserialize JSON strings into Python objects.
* Serialization and deserialization are just fancy terms for translating data between two different formats.
*When you serialize a Python object, like a dictionary, you're converting it into a JSON string so it can be easily sent across the web or saved to a file.
* Deserialization is the opposite—it's taking that JSON string and converting it back into a Python object so you can work with it in your code. In the notebook you're viewing, you can see examples of this using functions like `json.loads()` to parse JSON strings.
* Read from and write to .json files.
* Navigate and extract data from nested JSON structures.

##### Key Syntax Rules:
* Data is in key/value pairs.
* Keys must be strings in double quotes (").
* Values can be a string, a number, a boolean (true/false), an array (in square brackets []), or an object (in curly braces {}).
* Data is separated by commas.
* Curly braces {} hold objects.
* Square brackets [] hold arrays.

```python
import json
json_example_string = """
{
  "name": "John Doe",
  "age": 30,
  "isStudent": false,
  "courses": [
    {
      "title": "History",
      "credits": 3
    },
    {
      "title": "Math",
      "credits": 4
    }
  ],
  "address": null
}
"""

data = json.loads(json_example_string)#s removes the string and it is read  in dictionary
print(type(json_example_string))
print(data)
print(type(data))


# 2. Defining the JSON String
# What it does: This creates a multi-line Python string (using triple quotes """) containing raw text formatted as JSON.

# Key Details:

# In JSON, Boolean values are lowercase (false), and empty/null values are written as null.

# At this stage, Python views this variable strictly as a string (str), not as an interactive database or dictionary.
# 3. Deserializing with json.loads()
# What it does: This is where the magic happens.
# The loads() function (short for "load string") reads the raw text string (json_example_string) and translates
# it into a native Python data structure.
# How the Data Translates:
# JSON Element          Translated Python Object
# JSON Object { ... }   Python Dictionary dict
# JSON Array [ ... ]    Python List list
# JSON Boolean false     Python Boolean False
# JSON Null null         Python NoneType None

# Understanding loads vs. load (The "S" Rule)
# If you are parsing a String, use loads(). If you are reading directly from a File, use load().

```

```python
import json

python_user_data = {
    "id": 101,
    "name": "Jane Smith",
    "email": "jane.smith@example.com",
    "is_active": True,
    "courses": ["Biology", "Chemistry"]
}

with open("newfile.json", "w") as f:
    json.dump(python_user_data, f, indent=4)


with open("newfile.json", "r") as f:
    python_data = json.load(f)

print(python_user_data)
print("-"*20)
python_data

# the code demonstrates Serialization (saving data to a file) and Deserialization (reading data back from a file) 
# using an external file named newfile.json.

# # Writing to a File (json.dump)
# with open("newfile.json", "w") as f:: This opens (or creates) a file named newfile.json in write ("w") mode.
#     The with statement ensures the file is safely closed automatically once the block finishes running.

# json.dump(...): The dump() function (no "s" at the end!) takes your Python dictionary and 
# writes it directly as a formatted JSON stream into the file object f.

# indent = 4: This argument makes the output JSON file human-readable by adding clean indentation and 
# line breaks, rather than saving all the data in one long, messy line. During this process, Python's True is automatically converted to JSON's lowercase true inside the file.

# Reading from the File (json.load)
# with open("newfile.json", "r") as f:: This reopens the exact same file, but this time in read ("r") mode.

# json.load(f): The load() function (again, no "s"!) reads the raw JSON data straight from the file stream f and parses it 
# back into a native Python dictionary, which is saved into the new variable python_data.


# json.dumps() (Dump String)
# What it does: It serializes a Python object (like a dictionary) and converts it into a plain JSON string in your computer's memory.

# When to use it: When you don't want to save the data to a file yet, but instead need it as text—for example,
# to print it out, store it in a database, or send it over the web to an API.

# json.dump() (Dump to File)
# What it does: It serializes a Python object and writes it directly as a formatted stream into an external file (like the newfile.json in Cell 13).

# When to use it: When your goal is to physically save the data onto your computer's hard drive as a .json file.

# The "S" at the end stands for String.

# Use json.dumps() when you want a JSON String variable.

# Use json.dump() when you want to write directly to a File.
```

```python
import json

complex_json_string = """
{
  "companyName": "Tech Solutions Inc.",
  "products": [
    {
      "id": "P001",
      "name": "SuperWidget",
      "specs": {
        "storage": "256GB",
        "ram": "8GB"
      }
    },
    {
      "id": "P002",
      "name": "MegaDevice",
      "specs": {
        "storage": "512GB",
        "ram": "16GB"
      }
    }
  ]
}
"""
data = json.loads(complex_json_string)
# accessing the value of ram key of specs from the first product
print(data["products"][0]["specs"]["ram"])
company_name = data['companyName']
print("-----")

# 2. Get the name of the first product

product_name = data["products"][0]["name"]
print(product_name)

# 3. Get the RAM of the second product
# Access 'products' list (index 1), then 'specs' object, then 'ram' key
product_ram = data["products"][1]["specs"]["ram"]
print(product_ram)

```

Activity 1: `Convert User Data to JSON`

** Create a Python dictionary that represents your own user profile. It should include your name, an email, a list of hobbies, and 
a nested object for your address (city, zip code). Then, serialize it into a nicely formatted JSON string and print it.

```python
import json
myProfile = {
    "name" : "Sarita Shrestha",
    "email" : "sarita9@gmail.com",
    "hobbies" : ['Watching films', 'Skipping', 'Reading Books', 'Coding'],
    "address" : {
        "city" : "Lalitpur",
        "zip" : 5667,
    },
    "is_Nepali" : True
}
# serializing it into a nicely formatted JSON string

with open("myProfile.json", 'w') as f:
    json.dump(myProfile, f , indent = 4)

with open("myProfile.json") as f:
    content = json.load(f)
    
print("Original Dictionary:")
print(myProfile)
print("After storing and reading from the file-")
print(content)

```

Activity 2: `Extract and Manipulate Data`
* Given the JSON data below, parse it and perform the following tasks:
* Print the name of the manager.
* Loop through all employees and print their name and role.
* Calculate the total number of team members (employees).

```python
import json

team_data_json = """
{
  "department": "Engineering",
  "manager": "Alice Brown",
  "employees": [
    {
      "name": "Bob Johnson",
      "role": "Senior Developer"
    },
    {
      "name": "Charlie Davis",
      "role": "Junior Developer"
    },
    {
      "name": "Diana Miller",
      "role": "QA Engineer"
    }
  ]
}
"""

data= json.loads(team_data_json)


manager = data["manager"]
print(manager)



```

```python
name_of_manager= data["manager"]
print(name_of_manager)
```

```python
all_employee= data["employees"]
print(all_employee)
```

```python
count = 0
```

```python
for employee in all_employee:
    print("Employee Name:",employee["name"])
    print("Employee Role:",employee["role"])
    count+=1

print("Total number of team members:", count)
```

```python
import json

# --- Step 1: Assume config.json is created ---
data = {
  "theme": "dark",
  "notifications_enabled": True,
  "font_size": 14
}

with open('config.json', 'w') as f:
    json.dump(data, f, indent=4)

# --- Step 2: Read and Load ---
# Use a with statement for safe file handling
with open('config.json', 'r') as f:
    config = json.load(f)

# --- Step 3: Display ---
print("--- Initial Configuration ---")
print(config)

# --- Step 4: Update ---
# Your code here to modify the `config` dictionary
config['theme'] = "light"
config["font_style"] = "Times New Roman"

# --- Step 5: Write and Save ---
with open('config.json', 'w') as f:
    json.dump(config, f, indent=4) # Use indent for readability

# --- Step 6: Verify ---
print("\nConfiguration updated successfully!")
print(config)
```
