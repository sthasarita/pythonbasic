### 1. What is File Handling?
File handling refers to the process of creating, reading, updating, and deleting files using a program. It's essential for:
*   **Data Persistence:** Saving program data so it exists after the program finishes.
*   **Input:** Reading data from external sources (config files, datasets, user files).
*   **Output:** Storing results, logs, or generated reports.
*   **Data Transfer:** Exchanging information between different programs or systems.

###  Opening Files: `open()`

 The built-in `open()` function is used to open a file and returns a file object (also called a handle).


 `file_object = open('sample.txt', 'r')`

`'r'`  | **Read** (Default). Text mode.
`'w'`  | **Write**. Text mode. Creates file or **truncates** existing. | Truncates (empties)
`'a'`  | **Append**. Text mode. Creates file or appends to existing.
`'x'`  | **Exclusive Creation**. Text mode. Fails if file exists.

`'b'`  | **Binary** mode (append to mode, e.g., `'rb'`, `'wb'`).
`'t'`  | **Text** mode (Default, append to mode, e.g., `'rt'`, `'wt'`).
`'+'`  | **Update** (Reading and Writing, append to mode, e.g., `'r+'`, `'w+'`, `'a+'`)

```python
#for simplicity we are assigning the filename to a variable 
filename = "sample_one.txt"
print(f"Writing to {filename} using 'with'...")

with open(filename, "w") as f:
    f.write("Hello from Python!\n")
    f.write("This is the second line.\n")
    f.write("The 'with' statement is great.\n")
print(f"File automatically closed.")
```

```python
#reading the file sample_one.txt
with open(filename, 'r') as f:
    print(f.read())
```

```python
#here beacause we have write each sentence with \n and we are printing using print function so empty line is presented between the output line
with open(filename, 'r') as f:
    print(f.readline())
    print(f.readline())
```

```python
#readlines() converts every line to the list items and returns the list just like the split() method of string
with open(filename, 'r') as f:
    print(f.readlines())
    print(f.readlines())
```

```python
#f.seek() moves the file pointer to the character of the file whose actual position is passed in it 
with open(filename, 'r') as f:
    print(f.readline())
    f.seek(42)
    print(f.readline())
    print(f.readline())
```

```python
#f.seek() is mainly useful to move the file pointer back to the begining of the file i.e first position of the file so that after writing to the file we will be able to read from the initial postion of the file when file is open in w+ mode
with open(filename, "w+") as f:
    f.write("Hello from Python!\n")
    f.write("This is the second line.\n")
    f.write("The 'with' statement is great.\n")
    f.seek(0)
    print(f.read())
```

```python
#same as above 
filename = "sample_one.txt"
with open(filename, "a+") as f:
    f.write("Hello from Python!\n")
    f.write("This is the second line.\n")
    f.write("The 'with'.\n")
    f.seek(0)
    print(f.read())
```

```python
# writelines method writes each element to the file one by one so there is no need of explicit for loop
more_data = ['Python is versatile.\n' , 'File handling is key.\n']
with open(filename, "a+") as f:
    print("Appending using writelines():")
    f.writelines(more_data)
    f.seek(0)
    print(f.read())
    
```

```python
#writing to the file using write() function where we need to use loop as to store the content of the list 
output_filename = "output.txt"
data_to_write = ["First line.\n", "Second line.\n", "Third line without newline."]


# Method 1: Using write() in a loop
with open(output_filename, "w") as f:
    print(f"\nWriting using write() in a loop:")
    for line in data_to_write:
        num_chars = f.write(line)
        print(f"  Wrote {num_chars} characters.")

   
print("Finished writing with write().")

# Verify content
with open(output_filename, "r") as f:
    print("Content after write():")
    print(f.read())
```

```python
# using break statement after completion of reading files
with open("test.txt", "w") as f:
    f.write("this is first line\n")
    f.write("this is second line\n")
    f.write("this is third line")
with open("test.txt", "r") as f:
    while True:
        line = f.readline()
        if not line:
            break
        print(line)
```
