### 1. What are Sets?
 Sets are unordered collections of unique, immutable elements.
*   They do not store duplicate elements.
*   Elements must be immutable (like keys in dictionaries).
*   Sets themselves are mutable (you can add/remove elements).
*   There is no concept of order or index.
*   Very efficient for membership testing (`in`) and removing duplicates.

```python
# combining two set using |

set_a = {1,3,3}
set_b = {4,5,6,4}
union_set = set_a | set_b
union_set
```

```python
#combining two sets using union method
set_a = {1,3,3}
set_b = {4,5,6,4}
union_set_op = set_a.union(set_b)
union_set_op
```

```python
# Operator: |
# Method: .union()
union_set_op = set_a | set_b
union_set_meth = set_a.union(set_b)

print(f"Union (|): {union_set_op}")
print(f"Union (method): {union_set_meth}")
```

```python
# Operator: &
# Method: .intersection()
set_a = {1,5,3,23}
set_b = {1,5,32}
intersect_set_op = set_a & set_b
intersect_set_meth = set_a.intersection(set_b)
print(f"Intersection (&): {intersect_set_op}")
print(f"Intersection (method): {intersect_set_meth}")
print(set_a)
print(set_b)
```

```python
# update - add another set to existing set\
fruits = {'apple','mango','banana'}
fruits_2 = {'guava', 'watermelon'}
fruits.update(fruits_2)
print(fruits)
print(fruits_2)
```

```python
# The object in the update() method does not have to be a set, it can be any iterable object (tuples, lists, dictionaries etc.).
falful = {'apple','mango','banana'}
falful_2 = ['guava', 'watermelon']
falful.update(falful_2)
print(falful)
print(falful_2)
```

```python
# To remove an item in a set, use the remove(), or the discard() method.
# If the item to remove does not exist, remove() will raise an error.
even_numbers = {2,4,6,8,10,2}
print(even_numbers)
even_numbers.remove(4)
print(even_numbers)
```

```python
# If the item to remove does not exist, discard() will NOT raise an error.
even_numbers = {2,4,6,8,10,2}
print(even_numbers)
even_numbers.discard(5)
print(even_numbers)
```

```python
# pop method remove random item
even_numbers = {2,4,6,8,10}
print(even_numbers)
del_item = even_numbers.pop()
print(even_numbers)
print(del_item)
```

```python
# The clear() method empties the set
even_numbers = {2,4,6,8,10}
print(even_numbers)
even_numbers.clear()
print(even_numbers)
```

```python
# The del keyword will delete the set completely
even_numbers = {2,4,6,8,10}
print(even_numbers)
del even_numbers
print(even_numbers)
```

```python
# The union() method returns a new set with all items from both sets.
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}

set3 = set1.union(set2)
print(set3)
print(set1)
print(set2)
```

```python
# Join multiple sets with the union() method
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}
set3 = {"rita","gita"}
set4 = {"pokemon","doraemon"}
final_set = set1.union(set2,set3,set4)
print(final_set)
```

```python
# The intersection() method will return a new set, that only contains the items that are present in both sets.
set1 = {'dog','cat','elephant','mouse'}
set2= {'leopard','rhino','cat'}
set3 = set1.intersection(set2)
print(set3)
```

```python
# The difference() method will return a new set that will contain only the items from the first set that are not present in the other set.
set1 = {'dog','cat','elephant','mouse'}
set2= {'leopard','rhino','cat'}
set3 = set1.difference(set2)
print(set3)
```

```python
# The difference_update() method will keep the items from the first set that are not in the other set, but it will change the original set instead of returning a new set.
set1 = {'dog','cat','elephant','mouse'}
set2= {'leopard','rhino','cat'}
set3 = set1.difference_update(set2)
print(set3)
print(set1)
```

```python
# The symmetric_difference() method will keep only the elements that are NOT present in both sets.
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.symmetric_difference(set2)

print(set3)
```

```python
# The symmetric_difference_update() method will also keep all but the duplicates, but it will change the original set instead of returning a new set.
set1 = {'gita','sita','shyam'}
set2 = {'ram','shyam','hari'}
set3 = set1.symmetric_difference_update(set2)
print(set3)
print(set1)
print(set2)
```
