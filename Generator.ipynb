```python
# basic example
def numbers():
    yield 1
    yield 2
    yield 3
    yield 4

print(numbers())
```

```python
# Why use Generators?
# Suppose you want numbers from 1 to 1 billion.
# numbers = list(range(1000000000)) this occupies large memory space

def numbers():
    for i in range(1000000000):
        yield i
        
# Only one number exists in memory at a time.
# Memory usage is very small.
```

```python
# basic example
# We use the next() keyword to manually pull the next value from a generator. It tells the generator to resume execution, run the code until it hits the yield keyword, 
# return the value, and pause—all without losing track of local variables.
def demo():
    print("Start")
    yield 10

    print("Middle")
    yield 20

    print("End")
    yield 30
g = demo()
print(next(g))
print(next(g))
print(next(g))
```

```python
# Generator using loops
def even():
    for i in range(2, 11, 2):
        yield i

for num in even():
    print(num)
    
```

```python
# Practical Example – Reading a Large File

# Suppose a file contains 5 GB of text.

# Bad way Entire file goes into memory.
with open("sample_one.txt") as f:
    lines = f.readlines()
    

# Better way Only one line is loaded at a time.
def read_file(filename):
    with open(filename) as f:
        for line in f:
            yield line

for line in read_file("sample_one.txt"):
    print(line.strip())
```

```python
# Data Science Example 1 – Processing Large CSV Files
# Imagine a CSV with 10 million rows.
import csv

def load_data(file):
    with open(file) as f:
        reader = csv.DictReader(f)
        for row in reader:
            yield row

for row in load_data("employees.csv"):
    print(row["Name"])
    
# Only one row is in memory.
# This is commonly used in data pipelines.
```

```python
# Data Science Example 2 – Filtering Data

# Suppose your dataset has millions of employees.
employees = [
    {"name":"Ram","salary":50000},
    {"name":"Hari","salary":80000},
    {"name":"Sita","salary":90000},
]   
def high_salary(data):
    for person in data:
        if person["salary"] > 70000:
            yield person

for emp in high_salary(employees):
    print(emp)
```

```python
# Data Science Example 4 – Data Cleaning Pipeline

# Instead of storing intermediate results, generators can form a pipeline.
data = [
    "  Apple ",
    " Banana",
    "Mango  ",
    " Orange "
]

def clean_data(items):
    for item in items:
        yield item.strip().lower()

for fruit in clean_data(data):
    print(fruit)
```

```python
# Data Science Example 6 – Machine Learning Training Data
# Deep learning libraries often use generator-like objects to stream images from disk during training.
def training_data():
    dataset = [
        ("Image1", 0),
        ("Image2", 1),
        ("Image3", 0),
    ]

    for image, label in dataset:
        yield image, label

for image, label in training_data():
    print(image, label)
```
