```python
numbers = [1,2,4,5,6,8,9]
squares = []
for x in numbers:
    if x % 2 == 0:
        a = x ** 2
        squares.append(a)
print(squares)
        
```

```python
# list comprehension
numbers = [1,2,4,5,6,8,9]
squares = [x ** 2 for x in numbers if x % 2 == 0]
print(squares)
```

```python
evens = [x for x in range(10) if x % 2 == 0]
print(evens)
odds = [x for x in range(10) if x % 2 != 0]
print(odds)
```

```python
words = ["apple", "banana", "cherry","apricot"]
capital_words = [fruit.capitalize() for fruit in words]
print(capital_words)
```

```python
words = ["apple", "banana", "cherry","apricot"]
words = [fruit.capitalize() for fruit in words]
print(words)
```

```python
upper_fruits = [fruit.upper() for fruit in words]
print(upper_fruits)
```

```python
lower_fruits = [fruit.lower() for fruit in words]
print(lower_fruits)
```

```python
s = "Learning python programming"
vowels = [char for char in s if char in "aeiou"]
print(vowels)
consonants = [char for char in s if char not in "aeiou"]
print(consonants)
```

```python
words = ['apple', 'banana', 'cherry', 'apricot']
words = [fruit for fruit in words if 'a' in fruit]
print(words)
```

```python
#slicing in tuple
fruits = ("apple","banana","cherry","guava")
print(f"elements of fruits from index 1 to 3 = {fruits[1:4]}")
fruits = ("apple","banana","cherry","guava")
print(f"elements of fruits from index 0 to 3 = {fruits[:4]}")
fruits = ("apple","banana","cherry","guava")
print(f"elements of fruits from index 1 to 3 = {fruits[1:]}")
```

```python
# we can change the value of the list element which is the element of tuple
tup  = ("car", "jeep", "bike", [1,2])
tup[3][1]= 1
print(tup)
```

```python
# creating the new tuple with the value orange and add that tuple
fruits = ("apple","guava")
y = ("orange",)
fruits += y
print(fruits)
```
