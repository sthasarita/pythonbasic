```python
# count
numbers = [1,2,3,2,5,3]

count = numbers.count(2)
print(f"The number of times 2 appeared in {numbers} = {count}")
```

```python
# index
numbers = [1,2,3,2,5,3]

index = numbers.index(2)
print(f"The index at which 2 appeared first in {numbers} = {index}")
```

```python
# reverse
numbers = [1,2,3,2,5,3]
numbers.reverse()
print(numbers)
```

```python
# sort
numbers = [1,2,3,2,5,3]
numbers.sort()
print(numbers)
```

```python
# sort(descending)
numbers = [1,2,3,2,5,3]
numbers.sort(reverse = True)
print(numbers)
```

# slicing and iteration

```python
numbers = [0,1,2,3,4,5]
subset = numbers[::2]#similar to range functions
print(subset)
```

```python
numbers = [0,1,2,3,4,5]
subset = numbers[-3:-1]
print(subset)
```

```python
numbers = [0,1,2,3,4,5]
subset = numbers[:]
print(subset)
```

```python
numbers = [0,1,2,3,4,5]
subset = numbers[2:]
print(subset)
```

```python
numbers = [0,1,2,3,4,5]
subset = numbers[2:4]
print(subset)
```

```python
numbers = [0,1,2,3,4,5]
for num in numbers:
    print(num)
```

```python
fruits = ['apple', 'mango', 'orange']
for i in range(len(fruits)):
    print(f"fruits[{i}] = {fruits[i]}")

print(len(fruits))
```
