```python
#arithmetic operator +,-,%,*,/,//
# comparision operator ==, >=, <=, !=, <, >, 
# logical operator and or not
# bitwise operator & | ~ 
# assignment operator =, +=, -=

a = 15
b = 4

print(f"a = {a}, b = {b}")
print(f"Addition (a + b): {a + b}")           # Addition
print(f"Subtraction (a - b): {a - b}")        # Subtraction
print(f"Multiplication (a * b): {a * b}")     # Multiplication
print(f"Division (a / b): {a / b}")           # Division (returns float)
print(f"Floor Division (a // b): {a // b}")   # Floor Division (returns integer)
print(f"Modulus (a % b): {a % b}")            # Modulus (remainder)
print(f"Exponentiation (a ** b): {a ** b}")   # Exponentiation (a to the power b)



print("\n--- More Examples ---")

# Floor division vs regular division
print(f"10 / 3 = {10 / 3}")
print(f"10 // 3 = {10 // 3}")
# Modulus operator for remainder
print(f"17 % 5 = {17 % 5}") 

# Exponentiation
print(f"2 ** 3 = {2 ** 3}")  
print(f"5 ** 2 = {5 ** 2}")

# Negative numbers
print(f"-10 // 3 = {-10 // 3}")
print(f"-10 % 3 = {-10 % 3}")

# Comparison operators
x = 10
y = 5

print(f"x = {x}, y = {y}")
print(f"Equal to (x == y): {x == y}")
print(f"Not equal to (x != y): {x != y}")
print(f"Greater than (x > y): {x >= y}")

print(f"Less than (x < y): {x < y}")
print(f"Greater than or equal to (x >= y): {x >= y}")
print(f"Less than or equal to (x <= y): {x <= y}")


# Practical examples
age = int(input("Enter the age:"))
voting_age = 18
speed = int(input("Enter the speed:"))
speed_limit = 65
password = "python123"
entered_password = input("Enter the password:")

print(f"Can vote? (age >= voting_age): {age >= voting_age}")
print(f"Speeding? (speed > speed_limit): {speed > speed_limit}")
print(f"Password correct? (password == entered_password): {password == entered_password}")


result = (10 + 5) * 2 > 25
print(f"(10 + 5) * 2 > 25: {result}")


#bitwise
5 = 0101
3 = 0011
a= 5
b= 3

0*2**3  + 0*2**2  + 1*2**1  + 1*2**0

AND (&)

Compares each bit:

1 & 1 = 1
otherwise = 0


5 = 0101
& 3 = 0011
---------
       0001


2. OR (|)

If either bit is 1 → result is 1

5 = 0101
| 3 = 0011
------------
    = 0111  → 7


3. XOR (^)

Exclusive OR:

1 if bits are different
0 if same

5 = 0101
^ 3 = 0011
------------
    = 0110  → 6


4. NOT (~)

Flips all bits:

0 → 1
1 → 0

```
